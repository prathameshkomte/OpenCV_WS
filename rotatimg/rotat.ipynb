```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\prath\OneDrive\Pictures\Screenshots\Screenshot (70).png')
height, width = image.shape[:2]

# Divide by two to rototate the image around its centre
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, .5)

rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
```

```python
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')

plt.show()
```

```python
#Other Option to Rotate
img = cv2.imread(r'C:\Users\prath\OneDrive\Pictures\Screenshots\Screenshot (70).png')

rotated_image = cv2.transpose(img)

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image - Method 2')

plt.show()
```

```python
# Let's now to a horizontal flip.
flipped = cv2.flip(image, 1)

plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.title('Horizontal Flip')

plt.show()
```
