```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread(r'C:\Users\prath\Downloads\so bunny.png')
```

```python
B, G, R = image[10, 50] 
print (B, G, R)
print (image.shape)
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print (gray_img.shape)
print (gray_img[10, 50]) 
```

```python
gray_img[0, 0] 
```

```python
# OpenCV's 'split' function splites the image into each color index
B, G, R = cv2.split(image)

print (B.shape)

row, col = 1, 3
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(R, cv2.COLOR_BGR2RGB))
axs[0].set_title('Red')

axs[1].imshow(cv2.cvtColor(G, cv2.COLOR_BGR2RGB))
axs[1].set_title('Green')

axs[2].imshow(cv2.cvtColor(B, cv2.COLOR_BGR2RGB))
axs[2].set_title('Blue')

plt.show()
```

```python
row, col = 1, 3
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
merged = cv2.merge([B, G, R]) 
axs[0].imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB))
axs[0].set_title('Merged')

merged_blu = cv2.merge([B+100, G, R])
axs[1].imshow(cv2.cvtColor(merged_blu, cv2.COLOR_BGR2RGB))
axs[1].set_title('Merged with Blue Amplified')

merged_red = cv2.merge([B, G, R+100])
axs[2].imshow(cv2.cvtColor(merged_red, cv2.COLOR_BGR2RGB))
axs[2].set_title('Merged with Red Amplified')

plt.show()
```

```python
# Simply use 'imwrite' specificing the file name and the image to be saved
cv2.imwrite('merged_with_blue_amplified.jpg', merged_blu)
cv2.imwrite('merged_with_blue_amplified.png', merged_blu)
cv2.imwrite('merged_with_red_amplified.jpg', merged_red)
cv2.imwrite('merged_with_red_amplified.png', merged_red)
cv2.imwrite('merged.png', merged)
cv2.imwrite('merged.jpg', merged)
```
